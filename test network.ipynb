```python
import dataset
import numpy as np
import matplotlib.pyplot as plt
import torch

sample_rate_Hz = 10000.0
episode_duration_s = 1.0

ds = dataset.DatasetChirpingLIGOL(episode_duration_s, sample_rate_Hz, n_sources=1, episodes_per_epoch=64)
loader = torch.utils.data.DataLoader(ds, batch_size=8, num_workers=8, pin_memory=True)

for data, labels in loader:
    print(f"{data.shape=} {data.dtype=}\n{labels.shape=} {labels.dtype=}")
    plt.figure(figsize=(20, 4))
    plt.imshow(data[0, :, :].numpy(), origin='lower', aspect='auto')
    plt.show()
```
