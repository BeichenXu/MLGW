```python
import numpy as np

from tqdm.notebook import tqdm
import torch
import plotting
import networks
import sinc_source_utils as utils
```

# Source definition

let's start from a super simple source given by a sinc function parametrized by the vector $\theta$
$$ \phi(\theta) = sinc\left( \frac{2(t- \theta_1)}{\theta_0} \right) $$

```python
n_sources = 2
t = np.linspace(-5.0, 5.0, 128)
sources = utils.SincSource.sample(n_sources)
plotting.plot_multi(sources, t=t)
```

# Encode with net

```python
class SignalEncoder(torch.nn.Sequential):
    def __init__(self, max_sources: int, n_source_params: int):
        out_dims = max_sources * (n_source_params + 1)
        hidden_size = 32
        super().__init__(
            networks.conv_block(0, hidden_size),
            networks.conv_block(1, hidden_size),
            networks.conv_block(2, hidden_size),
            networks.conv_block(3, hidden_size),
            torch.nn.Flatten(start_dim=1),
            networks.linear_block(0, hidden_size=128),
            networks.linear_block(0, hidden_size=128),
            torch.nn.LazyLinear(out_features=out_dims),
        )
        self.n_sources = max_sources
        self.n_source_params = n_source_params

    def forward(self, x: torch.Tensor):
        x = super().forward(x)
        x = x.reshape(-1, self.n_sources, self.n_source_params + 1)
        return x
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
encoder = SignalEncoder(max_sources=n_sources, n_source_params=2).to(device)
print(encoder)
```

```python
encoder.eval()
signals, sources_target = utils.create_batch(batch_size=2, n_sources=n_sources, t=t)
params = encoder(signals.to(device))
sources_reconstructed = utils.get_sources_from(params)

for sources_t, sources_r in zip(sources_target, sources_reconstructed):
    plotting.plot_multi(sources_t, t=t, label="target")
    plotting.plot_multi(sources_r, t=t, label="reconst", new_fig=False)
```

```python
encoder.train()
optimizer = torch.optim.Adam(encoder.parameters(), lr=1e-3)
times = torch.as_tensor(t, dtype=torch.float32, device=device)

loss_log = []
for epoch in tqdm(range(25)):
    epoch_loss_log = []
    for batch in range(4*100):
        signals_targ, _ = utils.create_batch(batch_size=128, n_sources=n_sources, t=t)
        signals_targ = signals_targ.to(device)

        signals_rec = utils.decode(encoder(signals_targ), times).unsqueeze(1)
        loss = torch.nn.functional.mse_loss(signals_rec, signals_targ)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        epoch_loss_log.append(loss.item())
    loss_log.append(np.mean(epoch_loss_log))
    print(f"Epoch {epoch}: {loss_log[-1]}")
plotting.plot_loss(loss_log)
```
