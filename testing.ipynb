```python
import dataset
import numpy as np
import matplotlib.pyplot as plt
import torch
```

```python
N = 1_000_000
T = 1.0 / 10000.0
ds = dataset.DatasetChirping(N, T, examples_per_epoch=8)
loader = torch.utils.data.DataLoader(ds, batch_size=1, num_workers=8, pin_memory=True)
```

```python
for b, l in loader:
    print(b.shape, l)
    plt.plot(b[0])
    plt.show()
```
